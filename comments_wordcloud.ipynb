```python
# libraries
import pandas as pd
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from os import path
from PIL import Image
from wordcloud import wordcloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
# read dataset
df0 = pd.read_csv('dataset_e3.csv')
```

```python
df0.head(3)
```

```python
# filtro por proveedor
# 'Ediciones Tomodomo', 'Crocs', 'Ray-ban', 'Skechers', 'Pavco'
# Filter by city and suppliers

df_sup_city = df0[ 
                  (df0[ 'ciudad' ] == 'Gijón')
                & (df0[ 'proveedores' ]  == 'Ediciones Tomodomo')
                ]
```

```python

df_sup_city['comentarios'].to_csv('comments_sup_city_best.txt', index=False, sep=' ', header=None)
```

```python
# Wordcloud for the best suppliers
spanish_sw = stopwords.words("spanish")

comments_to_read = 'comments_sup_city_best.txt'

dirname = path.dirname( comments_to_read )

text = open( path.join( dirname, comments_to_read ) ).read()
#print(text)
#print( spanish_sw )

#cloud = wordcloud.WordCloud( stopwords= spanish_sw ).generate(text)
cloud = wordcloud.WordCloud( stopwords=spanish_sw ).generate(text)
plt.figure( figsize=( 18, 18 ), facecolor=None )
plt.imshow( cloud, interpolation='bilinear' )
plt.axis('off')
plt.tight_layout(pad=0)
```

```python
# The worst supplier in Gijón: 'Hollister'
df_worst_sup_city = df0[ 
                  (df0[ 'ciudad' ] == 'Gijón')
                & (df0[ 'proveedores' ]  == 'Hollister')
                ]
```

```python
df_worst_sup_city['comentarios'].to_csv('comments_sup_city_worst.txt', index=False, sep=' ', header=None)
```
