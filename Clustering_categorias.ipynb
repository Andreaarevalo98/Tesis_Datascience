```python

from  random import randint
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.cluster import AgglomerativeClustering
```

```python
df0 = pd.read_csv ('DATASET_BASE.csv')
```

```python
df0.head(3)
```

```python
df1= pd.read_csv ('Usuario.csv')
```

```python
df1.head(3)
```

```python
df_limpio = df1 [['Usuario', 'Edad', 'Gasto_mes', 'Tiempo_mes']]
```

```python
df_limpio.shape
```

```python
# Clustering herarquico
matriz = linkage (df_limpio, method='complete') 
plt.figure( figsize=(25,10))
dendrogram (matriz, leaf_rotation=90, leaf_font_size= 10);
plt.xlabel = ('Datos')
plt.ylabel = ('Distancia Euclidea')
plt.title =('Segmentacion de los usuarios')
```

```python
# Clustering herarquico
matriz = linkage (df_limpio, method='complete') 
plt.figure( figsize=(25,10))
dendrogram (matriz, leaf_rotation=90, leaf_font_size= 15, 
truncate_mode = 'lastp', p=7, show_leaf_counts = True, show_contracted = True);
plt.xlabel = ('Datos')
plt.ylabel = ('Distancia Euclidea')
plt.title =('Segmentacion de los usuarios')
```

```python
clusters= fcluster (matriz, 320, criterion= 'distance')
```

```python
clusters_serie = pd.Series (clusters)
```

```python
clusters_serie
```

```python
df_cluster = df1.copy()
```

```python
df_cluster ['ID_cluster'] = clusters_serie 
```
