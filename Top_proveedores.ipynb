## Top PROVEEDORES de la ciudad con mas ventas

```python

from  random import randint
import pandas as pd
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
df0 = pd.read_csv ('DATASET_BASE.csv')
```

```python
df0.head()
```

```python
df1= df0.copy()
```

```python
df1.head()
```

```python
df_valencia = df1[df1['Ciudad']=='Valencia']
```

```python
df_valencia.head(3)
```

```python
#df_valencia.groupby(['Proveedores']).sum()
```

```python
# Extraemos los atributos de interes para sacar proveedores mas vendidos de valencia 
df_atributos = df_valencia [['Fecha', 'Precio', 'Proveedores']]
```

```python
df_atributos.head()
```

```python
df_top_proveedores= df_atributos.groupby (['Proveedores']).sum().sort_values(by= 'Precio', ascending=False)
```

```python
df_top_proveedores.plot.bar(figsize = (15,6))
plt.title('Top proveedores de Valencia ')
plt.xlabel('Proveedores')
plt.ylabel('Ventas acumuladas ')
```

```python
# Tomamos la desicion de seleccionar los 10 proveedrores mas vendidos
```

```python
df_top_proveedores.head(10)
```

```python
df_top10= df_top_proveedores.iloc[:5]
```

```python
df_top10.plot.bar(figsize = (15,6))
plt.title('Top 5 proveedores de Valencia ')
plt.xlabel('Proveedores')
plt.ylabel('Ventas acumuladas ')
```

```python
df_top5 = df_atributos [ 
             (df_atributos['Proveedores']=='Ediciones Tomodomo')
             | (df_atributos['Proveedores']=='Valdemar')
             | (df_atributos['Proveedores']=='Cristasol')
             | (df_atributos['Proveedores']=='Ray-ban')
             | (df_atributos['Proveedores']=='Playmobile')
            
             ]
```

```python
df_top5.head(3)
```

```python
#pd.pivot_table(data=df5, index= ['Ciudad'], values= ['Precio'], columns= 'Fecha', aggfunc=np.sum)
```

```python
df_top5_fechas= pd.pivot_table(data=df_top5, index= ['Proveedores'], values= ['Precio'], columns= 'Fecha', aggfunc=np.sum)
```

```python
df_top5_fechas.plot.bar(figsize =(15,6))
plt.title('Ventas por temporadas del top 5 de proveedores de Valencia ')
plt.xlabel('Proveedores')
plt.ylabel('Ventas segregadas por temporada')
```

```python
df_top5_fechas.plot.bar(stacked = True, figsize =(15,6))
plt.title('Ventas por temporadas del top 5 de proveedores de Valencia ')
plt.xlabel('Proveedores')
plt.ylabel('Ventas segregadas por temporada')
```
